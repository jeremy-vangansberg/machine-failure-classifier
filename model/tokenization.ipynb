```python
import os
import numpy as np
import pandas as pd
import sentencepiece as spm
```

```python
# Read the Excel file
df = pd.read_excel("./data/historique_interventions_IA.xlsx", sheet_name="liste inter CATEGORISEES LIGHT")

# Rename columns to lowercase with underscores
df.columns = [title.lower().replace(" ", "_") for title in df.columns]

# Path to your text file for training the SentencePiece model
input_file = "./save_model/corpus_texte.txt"

# Preprocess and save the 'description' column to a text file
df["description"] = df["description"].replace(r'[.,()"\'\+]', ' ', regex=True)
df["description"].to_csv(input_file, header=False, index=False)

# Initialize SentencePiece processor
sp = spm.SentencePieceProcessor()

# Define the path to save the trained model
model_dir = "./save_model/"

# Train SentencePiece model and save it in the specified directory
spm.SentencePieceTrainer.train(input=input_file, model_prefix=os.path.join(model_dir, 'tokenizer_model'), vocab_size=733)
```
